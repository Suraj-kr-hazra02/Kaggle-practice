```python
#Write a Python program to read Assembly levels and their addresses and generate a symbole table.
symbole_table = {}

n = int(input("Enter the number of symbol: "))

for _ in range(n):
  lebel = input("Enter the lebel: ")
  address = input("Enter the address: ")
  symbole_table[lebel] = address

print("\n----- SYMBOL TABLE -----")
print("Lebel\tAddress")

for lebel, address in symbole_table.items():
  print(f"{lebel}\t{address}")
```

```python
#Write a python program to simulate the location counter while processing assembly instructions.
locctr = int(input("Enter starting address:"))
n = int(input("Enter number of instructions:"))

print("\nInstruction\tAddress")

for _ in range(n):
    instruction = input("Enter instruction:")
    print(f"{instruction}\t\t{locctr}")
    locctr += 4

print("\nFinal Location Counter =", locctr)
```

```python
#Write a python program to implement pass 1 of a two pass a assembler 
#by generating the symbole table and intermediate code.

symbol_table = {}
intermediate_code = []

# Starting address
lc = int(input("Enter the starting address: "))

# Number of instructions
n = int(input("Enter the number of instructions: "))

print("\nEnter instructions in the format:")
print("Label Opcode Operand")
print("(Use - if there is no label or operand)\n")

for i in range(n):
    label, opcode, operand = input().split()

    # Enter label into symbol table
    if label != "-":
        symbol_table[label] = lc

    # Store intermediate code
    intermediate_code.append([lc, opcode, operand])

    # Increment location counter
    lc += 1

# Display Symbol Table
print("\n------ SYMBOL TABLE ------")
print("Symbol\tAddress")
print("--------------------------")

for symbol, address in symbol_table.items():
    print(f"{symbol}\t{address}")

# Display Intermediate Code
print("\n------ INTERMEDIATE CODE ------")
print("LC\tOpcode\tOperand")
print("-------------------------------")

for code in intermediate_code:
    print(f"{code[0]}\t{code[1]}\t{code[2]}")
```
