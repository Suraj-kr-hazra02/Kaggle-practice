```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()

print("Dataset Shape:", digits.data.shape)
print("Image Shape:", digits.images.shape)
print("Target Label:", digits.target_names)

plt.figure(figsize=(4,4))
plt.imshow(digits.images[2], cmap='gray')
plt.title(f"Label: {digits.target[0]}")
plt.axis('off')
plt.show()
```

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()

fig, axes = plt.subplots(2, 5, figsize=(10, 5))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f"Label: {digits.target[i]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
#write a python program to import iris dataset from speakitlearn and convert it into pandas data frem and display the shape of the dataset.
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)

df["target"] = iris.target

print("Shape of the dataset:", df.shape)

print("\nFirst 5 rows:")
print(df.head())
```
